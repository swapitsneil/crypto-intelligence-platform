```python
import numpy as np
import pandas as pd
import requests
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
API_KEY = os.getenv("API_KEY")

url = "https://api.coingecko.com/api/v3/coins/markets"
```

```python
params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 100,
    "page": 1,
    "sparkline": False
}

headers = {
    "x-cg-demo-api-key": API_KEY
}
```

```python
response = requests.get(url, params=params, headers=headers)

print(response)
print(response.status_code)
print(response.text)
```

```python
data = response.json()

df = pd.DataFrame(data)

print(df.shape)

df.head()
```

```python
keep_cols = [
    "id",
    "symbol",
    "name",
    "current_price",
    "market_cap",
    "market_cap_rank",
    "total_volume",
    "price_change_percentage_24h",
    "circulating_supply",
    "last_updated"
]

silver_df = df[keep_cols].copy()

silver_df.head()
```

```python
silver_df.info()
```

```python
silver_df["last_updated"] = pd.to_datetime(silver_df["last_updated"])
```

```python
silver_df.info()
```

```python
from datetime import datetime

silver_df["fetch_time"] = datetime.now()
```

```python
silver_df.head()
```

```python
silver_df.isnull().sum()
```

```python
silver_df["price_change_percentage_24h"] = (silver_df["price_change_percentage_24h"].fillna(0))
```
